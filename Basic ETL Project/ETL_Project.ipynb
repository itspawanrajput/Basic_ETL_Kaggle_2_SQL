# 2. Transform

```python
#reading the file and handling the missing values and null values
import pandas as pd
df = pd.read_csv('orders.csv',na_values=['Not Available','unknown'])
df['Ship Mode'].unique()
```

```python
#renaming the columns and make them lower case and replace space with underscore

df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.columns
```

```python
#devive new columns discount, sale_price, profit
df['discount']=df['list_price']*df['discount_percent']*.01
df['sale_price']=df['list_price']-df['discount']
df['profit']=df['sale_price']-df['cost_price']
df.head(5)
```

```python
df.info()
```

```python
df['order_date']=pd.to_datetime(df['order_date'],format="%Y-%m-%d")  ##converting the order_date to datetime format
```

```python
df.info()
```

```python
df.head()
```

```python
df.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)  #dropping the columns
df.head()
```

```python
df.head()
```

# 3. Load Data

## Load data into  SQL

```python
#loading the data into the database
import sqlite3
conn = sqlite3.connect('orders.db')
df.to_sql('orders',conn,if_exists='replace',index=False)
conn.close()
```

```python
#read the data from the database
import sqlite3
conn = sqlite3.connect('orders.db')
data = pd.read_sql('SELECT * FROM orders',conn)
data.head()
```

```python
#total sales, profit, discount
data = pd.read_sql('SELECT SUM(sale_price) as total_sales, SUM(profit) as total_profit, SUM(discount) as total_discount FROM orders',conn)
data.head()
```

```python
# find top 10 highest revenew generating products
data = pd.read_sql('SELECT product_id, SUM(sale_price) as total_sales FROM orders GROUP BY product_id ORDER BY total_sales DESC LIMIT 10',conn)
data.head()
```

```python
# find top 10 highest revenew generating cities
data = pd.read_sql('SELECT city, SUM(sale_price) as total_sales FROM orders GROUP BY city ORDER BY total_sales DESC LIMIT 10',conn)
data.head()
```

```python
# find top 5 highest selling products in each region    
data = pd.read_sql('SELECT region, product_id, SUM(sale_price) as total_sales FROM orders GROUP BY region, product_id ORDER BY region, total_sales DESC',conn)
data.groupby('region').head(5)
data.head()
```

```python
# for each category find the month with the highest sales
data = pd.read_sql('SELECT category, strftime("%m",order_date) as month, SUM(sale_price) as total_sales FROM orders GROUP BY category, month ORDER BY category, total_sales DESC',conn)
# get the highest sales month for each category
result = data.groupby('category').head(1)
print("\nHighest sales month for each category:")
print(result)
```
